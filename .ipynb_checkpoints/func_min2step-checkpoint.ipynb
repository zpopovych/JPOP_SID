## Two step minimization function

On the first stage we perform global minimization with the constraints and then local minimization around the all minimizer candidates. Major steps are:
1. Set Mosek as glabal optimizer tool (use JuMP.optimizer_with_attributes function from the JuMP.jl package to group an optimizer constructor with the list of attributes)
https://jump.dev/JuMP.jl/stable/reference/models/#JuMP.optimizer_with_attributes
2. Find candidate minimizers (use MomentTools.minimize function) <br>
http://www-sop.inria.fr/members/Bernard.Mourrain/software/MomentTools/code/2.optimization.html#MomentTools.minimize
3. Loop local minimization using NLopt over all candidates 
 - convert polynomial expression of objective to function to be minimized
 - create NLopt model
 - set algorithm for NLopt
 - declare variables for model 
 - convert polynomial expression of constraint to function to be minimized
 - register objective and constraint with the model
 - perform optimization of the model 
 - substitute solition back to the polynomial expression of objective
 - choose the best minimizer 

```julia
using JuMP         # to set Mosek as optimizer 
using MosekTools   # to use as optimizer
using MomentTools  # to find minimizer candidates 
using NLopt        # to find local optimum
```

```julia
function min_SLSQP(obj, constr)
    # extract valiables from the objective
    vars = variables(obj)
    
    function g(a...)
        # Converting polynomial expression of objective to function to be minimized
        obj(vars => a)
    end
    
    function e(a...)
        # Converting polynomial expression of constraint to function to be minimize
        constr(vars => a)
    end
       
    # Create NLopt model
    model = Model(NLopt.Optimizer)

    # Set algorithm 
    set_optimizer_attribute(model, "algorithm", :LD_SLSQP) 
    
    # Set variables
    @variable(model, y[1:length(vars)]);

    # Register constraint
    register(model, :e, length(y), e; autodiff = true)
    
    @NLconstraint(model, e(y...) == 0)

    # Register objective
    register(model, :g, length(y), g; autodiff = true)
    @NLobjective(model, Min, g(y...))
    
    # Set guess
    guess = ones(length(y))
    for (var, init_val) in zip(y, guess)
        set_start_value(var, init_val)
    end

    # Call JuMP optimization function
    JuMP.optimize!(model)

    solution = vars => map(value, y)

    return(solution)

end

function min_SLSQP(obj)
    # extract valiables from the objective
    vars = variables(obj)
    
    function g(a...)
        # Converting polynomial expression of objective to function to be minimized
        obj(vars => a)
    end
    
    function e(a...)
        # Converting polynomial expression of constraint to function to be minimize
        constr(vars => a)
    end
       
    # Create NLopt model
    model = Model(NLopt.Optimizer)

    # Set algorithm 
    set_optimizer_attribute(model, "algorithm", :LD_SLSQP) 
    
    # Set variables
    @variable(model, y[1:length(vars)]);

    # Register constraint
    register(model, :e, length(y), e; autodiff = true)
    
    @NLconstraint(model, e(y...) == 0)

    # Register objective
    register(model, :g, length(y), g; autodiff = true)
    @NLobjective(model, Min, g(y...))
    
    # Set guess
    guess = ones(length(y))
    for (var, init_val) in zip(y, guess)
        set_start_value(var, init_val)
    end

    # Call JuMP optimization function
    JuMP.optimize!(model)

    solution = vars => map(value, y)

    return(solution)

end
```

```julia
using TSSOS

function min2step(obj, constr)
    # obj - is objective function
    # constr - one constraint in the form of equation
    
    # Perform global minimization with TSSOS package
    opt,sol,data = tssos_first([obj, constr], variables(obj), 2, numeq=1, solution=true, QUIET = true);
    
    # execute higher levels of the TSSOS hierarchy
    iter = 1
    best_opt = opt
    
    while !isnothing(opt)
        opt,sol,data = tssos_higher!(data, solution=true, QUIET = true)
        if !isnothing(opt) & (iter < 5)
            iter += 1
            if abs(opt) < abs(best_opt)
                best_opt = opt
            end
        end
    end
    
    # extract valiables from the objective
    vars = variables(obj)
    
    function g(a...)
        # Converting polynomial expression of objective to function to be minimized
        obj(vars => a)
    end
    
    function e(a...)
        # Converting polynomial expression of constraint to function to be minimize
        constr(vars => a)
    end
       
    # Create NLopt model
    model = Model(NLopt.Optimizer)

    # Set algorithm 
    set_optimizer_attribute(model, "algorithm", :LD_SLSQP) 
    
    # Set variables
    @variable(model, y[1:length(vars)]);

    # Register constraint
    register(model, :e, length(y), e; autodiff = true)
    
    @NLconstraint(model, e(y...) == 0)

    # Register objective
    register(model, :g, length(y), g; autodiff = true)
    @NLobjective(model, Min, g(y...))
    
    # Set guess
    guess = sol
    for (var, init_val) in zip(y, guess)
        set_start_value(var, init_val)
    end

    # Call JuMP optimization function
    JuMP.optimize!(model)

    solution = vars => map(value, y)

    return(solution)
end  
```

### Example

Consider some random Lindblad evolution

```julia
include("LiPoSID.jl")
seed = 100
w = 0.0
t = [0:0.01:1;]

using QuantumOptics
basis = NLevelBasis(2)
ρ = LiPoSID.rand_Linblad_w_noise(basis, seed, w, t)

using Plots
plot(LiPoSID.bloch(ρ)')
```

Build Pade and Simpson objectives fro this time series

```julia
using DynamicPolynomials

@polyvar x[1:4]
H_symb = [ 1.0 * x[1]              x[3] + im * x[4]
           x[3] - im * x[4]        x[2]             ]

@polyvar a[1:2, 1:2]
@polyvar b[1:2, 1:2]
J_symb = 1.0 * a + im * b

obj_pade = LiPoSID.pade_obj(ρ, t, H_symb, J_symb)
obj_simp = LiPoSID.simpson_obj(ρ, t, H_symb, J_symb);
```

Build Kraus objective and constraint

```julia
@polyvar a1[1:2, 1:2]
@polyvar b1[1:2, 1:2]
K1_symb = 1.0 * a1 + im * b1

@polyvar a2[1:2, 1:2]
@polyvar b2[1:2, 1:2]
K2_symb = 1.0 * a2 + im * b2

obj_kraus, constr_kraus = LiPoSID.kraus_obj_constr(ρ, K1_symb, K2_symb);
```

```julia
solution = min2step(obj_kraus, constr_kraus)
obj_kraus(solution)
```

```julia
solution = min_SLSQP(obj_kraus, constr_kraus)
obj_kraus(solution)
```

```julia
K1_sid = convert(Matrix{ComplexF64}, subs(K1_symb, solution))
K2_sid = convert(Matrix{ComplexF64}, subs(K2_symb, solution))
```

```julia
time_steps = length(t)
ρ_sid_kraus = LiPoSID.timeevolution_kraus(time_steps, ρ[1], [K1_sid, K2_sid])
plot(LiPoSID.bloch(ρ_sid_kraus)')
```
